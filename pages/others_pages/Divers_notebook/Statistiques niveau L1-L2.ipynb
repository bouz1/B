## Bar chart vs histogram vs density 

| Feature                | **Bar Chart**                                         | **Histogram**                                      | **Density Plot**                                  |
|------------------------|-------------------------------------------------------|----------------------------------------------------|---------------------------------------------------|
| **Purpose**             | Visualize categorical data or count comparisons.      | Visualize the distribution of continuous data.     | Show a smoothed representation of continuous data distribution. |
| **Data Type**           | Categorical data (discrete).                          | Continuous data (numeric).                         | Continuous data (numeric).                        |
| **X-axis**              | Categories (e.g., gender, regions).                   | Data bins (ranges of values).                      | Values of the continuous variable.               |
| **Y-axis**              | Frequency or count of each category.                  | Frequency or count of data within bins.            | Probability density (the smooth curve's height).  |
| **Appearance**          | Discrete bars for each category.                      | Bars representing frequency in specific bins.      | Smooth, continuous curve.                        |
| **Use Case**            | Comparing counts for distinct categories.             | Showing the frequency distribution of continuous data. | Understanding the overall distribution shape and density of continuous data. |
| **Best For**            | Categorical data analysis.                           | Distribution of numerical data (e.g., ages, test scores). | Distribution of numerical data with a smooth, continuous form. |
| **Examples**            | Number of products sold by type, survey response categories. | Age distribution of a population, test scores.     | Visualizing the normal distribution, income distribution. |


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 

arr=np.random.normal(loc=10,scale= 3, size= 100)

fig, axs=plt.subplots(1,4, figsize=(18,3))
pd.Series(arr.astype(int)).value_counts().sort_index().plot.bar(ax=axs[0], title="bar chart")

bins= [3,6,8,9,10,11,12,14,17]
hist, _= np.histogram(arr, bins=bins)
x= [(bins[i],bins[i+1]) for i in range(len(bins)-1)]
x2= [str(v) for v in x]
axs[1].bar(x2, hist)
axs[1].set_title("histogram (Freq)")
#axs[1].set_xticks(rotation=70)

axs[2].set_xticklabels(x2, rotation=90)  # Rotate labels for axs[1] only
freq= hist/np.sum(hist)
axs[2].bar(x2,freq)
axs[2].set_title("histogram (relative freq)")
#axs[2].set_xticks(rotation=70)

axs[2].set_xticklabels(x2, rotation=90)  # Rotate labels for axs[1] only

# this is the definition of the density 
density= [f/(h[1]-h[0]) for h, f in zip(x,freq)]

axs[3].bar(x2,density)
axs[3].set_title("Density")
axs[3].set_xticklabels(x2, rotation=90)  # Rotate labels for axs[1] only

plt.show()
```

```python
xx= [(h[1]-h[0]) for h in x]
xxx=np.cumsum(xx)
yy= density
print(np.trapezoid(y= yy,x= xxx), "close to one")
```

```python
# Integral with rectagles are
print(np.sum([(xv*yv) for xv,yv in zip(xx, yy)]), "equal 1")
```
